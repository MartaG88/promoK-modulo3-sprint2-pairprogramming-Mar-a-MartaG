2. Ejercicios intervalo de confianza. Al igual que en el caso del ejercicio de la clase invertida, piensa un motivo por el que ser铆a interesante calcular estos intervalos de confianza para la extracci贸n de insights.

Intervalo de Confianza para la Esperanza de Vida

Contexto: Trabajas en una organizaci贸n de salud global y deseas estimar el intervalo de confianza al 95% para la esperanza de vida promedio de cada continente.

Nota: Cuidado porque puede que os salgan algunos nulos, genera un c贸digo para que no salgan nulos en los intervalos de confianza.

Intervalo de Confianza para el PIB

Contexto: Eres una economista y quieres calcular el intervalo de confianza al 90% para el Producto Interno Bruto (PIB) de los continentes.

```python
# lo primero que tenemos que hacer es calcular la estimaci贸n puntual

media_le = df_le["life_expectancy"].mean()


# despu茅s vamos a calcular el error est谩ndar utilizando el m茅todo "sem" de la librer铆a stats
error_le = stats.sem(df_le["life_expectancy"])

# definimos el nivel de confianza (95% en este caso)
nivel_confianza_le = 0.95

# calculamos los grados de libertad de la muestra. Recordad que debemos restar el total de datos que tenemos -1. 
grados_libertad_le = len(df_le["life_expectancy"]) - 1

# calculamos el valor cr铆tico de la muestra
valor_critico_le = stats.t.ppf((1 + nivel_confianza_le) / 2, df=grados_libertad_le)

# calculamos el intervalo de confianza
limite_inferior_le = media_le - valor_critico_le * error_le
limite_superior_le = media_le + valor_critico_le * error_le

print("Intervalo de Confianza para la Esperanza de Vida:")
print(f"Media Muestral: {np.round(media_le, 2)}")
print(f"Error Est谩ndar: {np.round(error_le, 2)}")
print(f"Nivel de Confianza: {nivel_confianza_le}")
print(f"Valor Cr铆tico: {np.round(valor_critico_le, 2)}")
print(f"Intervalo de Confianza: ({np.round(limite_inferior_le, 2)}, {np.round(limite_superior_le, 2)})")
```
